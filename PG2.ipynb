```python
# Iterar sobre cada linha da tabela1
for index, row in tabela_unica.iterrows():
    nome_parlamentar = row['txNomeParlamentar']

    # Verifica se a string contém "LIDERANÇA DO" e extrai o texto após isso
    if "LIDERANÇA DO" in nome_parlamentar:
        partido_extraido = nome_parlamentar.replace("LIDERANÇA DO ", "").strip()

        # Verifica se partido_extraido está em siglaPartido
        match = df_parlamentares_filtrados[df_parlamentares_filtrados['siglaPartido'] == partido_extraido]

        # Se não encontrar, verifica em nomePartido
        if match.empty:
            match = df_parlamentares_filtrados[df_parlamentares_filtrados['nomePartido'] == partido_extraido]

        # Se encontrou correspondência, atualiza ideCadastro
        if not match.empty:
            tabela_unica.at[index, 'ideCadastro'] = match.iloc[0]['id']

# Converter ideCadastro para inteiro, preenchendo NaN temporariamente com 0
tabela_unica['ideCadastro'] = tabela_unica['ideCadastro'].fillna(0).astype(int)
tabela_unica.head()
```

```python
import time
import re

cnpj_url_base = 'https://open.cnpja.com/office/'

# Número máximo de requisições
max_requisicoes = 20
limite_por_minuto = 5

resultados = []

for i, cnpj_value in enumerate(df_cnpj_unico['txtCNPJCPF'][:max_requisicoes]):
  cnpj_limpo = re.sub(r"\D", "", cnpj_value)
  print(cnpj_limpo)
  url = cnpj_url_base + str(cnpj_limpo)
  response = requests.get(url)

  if response.status_code == 200:
    data = response.json()
    # Extraindo as informações relevantes
    resultado = {
      "cnpj": cnpj_limpo,
      "nome_fantasia": data.get("alias"),
      "data_fundacao": data.get("founded"),
      "status": data.get("status", {}).get("text"),
      "nome_empresa": data.get("company", {}).get("name"),
      "capital_social": data.get("company", {}).get("equity"),
      "natureza_juridica": data.get("company", {}).get("nature", {}).get("text"),
      "atividade_central": data.get("mainActivity", {}).get("text"),
      "endereco": f"{data.get('address', {}).get('street', '')}, {data.get('address', {}).get('number', '')} - {data.get('address', {}).get('city', '')} ({data.get('address', {}).get('state', '')})",
      "email": data.get("emails", [{}])[0].get("address") if data.get("emails") else None,
      "telefone": f"({data.get('phones', [{}])[0].get('area', '')}) {data.get('phones', [{}])[0].get('number', '')}" if data.get("phones") else None
    }
    resultados.append(resultado)
  else:
    print(f"Erro ({response.status_code}): {cnpj_value}")

  if (i + 1) % limite_por_minuto == 0 and (i + 1) < max_requisicoes:
    print("Aguardando 60 segundos para respeitar o limite de requisições...")
    time.sleep(60)

df_empresas = pd.DataFrame(resultados)
df_empresas.head()
```
